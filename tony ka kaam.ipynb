```python
def remove_unnamed(df):
    """
    Remove unnamed columns from the given dataframe if unnamed columns exist
    """
    if "Unnamed: 0" in df.columns:
        del df["Unnamed: 0"]
    return df


for df in [
    rate_of_coverage_model,
    market_share_model,
    rate_of_coverage,
    charecteristics,
    demand,
    total_purchase,
    additional,
    first_time,
    rate_of_coverage_segment,
    segments_size,
    market_share,
    marketing_costs,
    marketing_retail,
    marketing_end,
]:
    df = remove_unnamed(df)
```
