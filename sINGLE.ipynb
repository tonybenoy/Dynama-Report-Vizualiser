```python
import pandas as pd

marketing_end_df = pd.read_csv("marketing_end_df.csv")
marketing_end_df.drop("Unnamed: 0", axis=1, inplace=True)
marketing_retail_df = pd.read_csv("marketing_retail_df.csv")
marketing_retail_df.drop("Unnamed: 0", axis=1, inplace=True)
marketing_cost_df = pd.read_csv("marketing_costs_df.csv")
marketing_cost_df.drop("Unnamed: 0", axis=1, inplace=True)
charecteristics_df = pd.read_csv("charecteristics_df.csv")
charecteristics_df.drop("Unnamed: 0", axis=1, inplace=True)
additional_df = pd.read_csv("additional_df.csv")
additional_df.drop("Unnamed: 0", axis=1, inplace=True)
demand_df = pd.read_csv("demand_df.csv")
demand_df.drop("Unnamed: 0", axis=1, inplace=True)
first_time_df = pd.read_csv("first_time_df.csv")
first_time_df.drop("Unnamed: 0", axis=1, inplace=True)
rate_of_coverage_df = pd.read_csv("rate_of_coverage_df.csv")
rate_of_coverage_df.drop("Unnamed: 0", axis=1, inplace=True)
rate_of_coverage_segment_df = pd.read_csv("rate_of_coverage_segment_df.csv")
rate_of_coverage_segment_df.drop("Unnamed: 0", axis=1, inplace=True)
segment_size_df = pd.read_csv("segments_size_df.csv")
segment_size_df.drop("Unnamed: 0", axis=1, inplace=True)
total_purchases_df = pd.read_csv("total_purchase_df.csv")
total_purchases_df.drop("Unnamed: 0", axis=1, inplace=True)
market_share_df = pd.read_csv("market_share_df.csv")
market_share_df.drop("Unnamed: 0", axis=1, inplace=True)
sales_df = pd.read_csv("sales_df.csv")
```

```python
n_df=pd.melt(
    marketing_cost_df,
    id_vars=["Period", "Company"],
    value_vars=["1", "2", "3", "4", "5", "6"],
)
n_df.rename(columns={"variable": "Model","value":"Market Share"}, inplace=True)
charecteristics_df
```

```python
charecteristics_df["Model_Split"]=charecteristics_df["Model"].str.extract('(\d+)')
charecteristics_df.rename(columns={"Model":"Model Name","Model_Split": "Model"}, inplace=True)
```

```python
marketing_end_df.rename(columns={"Cost": "Marketing End User"}, inplace=True)
```

```python
n_df=n_df.merge(marketing_end_df, on=["Period", "Company"], how="left")
```

```python
marketing_retail_df.rename(columns={"Cost": "Marketing Retailer"}, inplace=True)
n_df = n_df.merge(marketing_retail_df, on=["Period", "Company"], how="left")
```

```python
n_df = n_df.merge(charecteristics_df, on=["Period", "Model","Company"], how="left")
```

```python
demand_df["Model_Split"]=demand_df["Model"].str.extract('(\d+)')
demand_df.drop("Model", axis=1, inplace=True)
demand_df.rename(columns={"Model_Split": "Model"}, inplace=True)
n_df = n_df.merge(demand_df, on=["Period", "Model","Company"], how="left")
```

```python
R_C =pd.melt(
    rate_of_coverage_df,
    id_vars=["Period", "Company"],
    value_vars=["1", "2", "3", "4", "5", "6"],
)
```

```python
R_C.rename(columns={"variable": "Model","value":"Rate of Coverage"}, inplace=True)
R_C
```

```python
n_df = n_df.merge(R_C, on=["Period", "Model","Company"], how="left")
```

```python
sales_df["Model_Split"]=sales_df["Model"].str.extract('(\d+)')
sales_df.rename(columns={"Model":"Model Name","Model_Split": "Model"}, inplace=True)
```

```python
n_df = n_df.merge(sales_df, on=["Period", "Model"], how="left")
```

```python
n_df.to_csv("n_df.csv", index=False)
```
